# Spotify song data analysis

```python
###
# Author: Sriram Vadlamani
# Start Date: 12th February 2020
# Last edit: 2nd August 2020
###
import os
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the dataset
path = "/home/sriram/Documents/repo/datarecon/SpotifySongAnalysis/SpotifySongAttribs.csv"
data = pd.read_csv(path)
```

```python
# Generating an average column for the given columns:
# Let's call it the Song factor
'''
acousticness, danceability, energy, instrumentalness, speechiness
'''
data['Song factor'] = data[['acousticness', 'danceability', 'energy', 'instrumentalness', 'speechiness']].mean(axis=1)
```

```python
# Plotting the song factor against the index of the song.
data.plot(x='Unnamed: 0', y='Song factor', color='c')
plt.xlabel('Song number')
plt.ylabel('Song factor')
```

```python
# Now, since the given song factor's value is already between zero and one, let's perform an FFT on it.
dt = 1
t = data['Unnamed: 0'].to_numpy() # The index of the song vector
n = len(data['Unnamed: 0'])
SongFactorArray = data['Song factor'].to_numpy()
fhat = np.fft.fft(SongFactorArray, n)
PSD = fhat * np.conj(fhat) / n # The power spectral density.
freq = (1/(dt * n)) * np.arange(n)
L = np.arange(1, np.floor(n/2), dtype='int')
```

```python
# Now plotting it.
plt.plot(freq[L], PSD[L], color='c', LineWidth='1.5', label='PSD')
plt.xlabel('Song Factor')
plt.ylabel('Song Factor density')
```

```python
'''
We can see that the Songs with song factor 0.0 < x < 0.1 have the most density, meaning most songs in the database
have a very low song factor. In other words, most popular songs on spotify are not very good.
'''
```
